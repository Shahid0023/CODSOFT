```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
#Loading Dataset  in variable
cr_data = pd.read_csv('/content/creditcard.csv')
```

```python
#Showing 1st 5 rows
cr_data.head()
```

```python
#Showing last 5 rows
cr_data.tail()
```

```python
#Dataset information
cr_data.info()
```

```python
#Dataset description
cr_data.describe()
```

```python
#checking for null values
cr_data.isnull().sum()
```

```python
#Class distribution 0 reperesents normal transactions and 1 reperesents fraud transactions
cr_data['Class'].value_counts()
```

```python
# As we can see the dataset is very un-balanced here we take a sample data from the normal count and store is in a variable
normaltransaction = cr_data[cr_data.Class==0]
fraudtransaction = cr_data[cr_data.Class==1]
```

```python
#shape attribute of dataset shows total no of rows and columns in the dataset
print(normaltransaction.shape)
print(fraudtransaction.shape)
```

```python
# statistical veiw of the dataset
normaltransaction.Amount.describe()
```

```python
fraudtransaction.Amount.describe()
```

```python
#The dataset is very un-balanced so we will take a sample data from normaltransaction dataset for training
# we will take same numbers of rows from normaltransaction as fraudtransaction are 492 but randomly
normal_sample = normaltransaction.sample(n=492)
```

```python
#lets check the sample dataset / you can see the data is ramdom
normal_sample.head()
```

```python
#you can see by 1st column the data is ramdom
normal_sample.tail()
```

```python
#Same row count as fraud data
print(normal_sample.shape)
```

```python
#now we will concat both of the datasents into one sinlge dataset to go for training
final_dataset = pd.concat([normal_sample,fraudtransaction],axis=0)
```

```python
print(final_dataset.shape)
```

```python
#now check for value count by class again
final_dataset['Class'].value_counts()
```

```python
final_dataset.groupby('Class').mean()
```

```python
#Splitting data into X and Y / Target data and dependancies
X = final_dataset.drop(columns='Class',axis = 1)
Y = final_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Lets train our dataset
X_train, X_test, Y_train, Y_test  = train_test_split(X,Y,test_size=0.2 , stratify=Y, random_state=2 )
```

```python
#now print the shape
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model training with logistic regression
cModel = LogisticRegression()
```

```python
cModel.fit(X_train,Y_train)
```

```python
#Evaluation & accuracy core checking with train data
X_tainPred= cModel.predict(X_train)
X_train_AccScore = accuracy_score(X_tainPred,Y_train)
```

```python
print('Accuracy with Training data : ', X_train_AccScore)
```

```python
#Evaluation & accuracy core checking with test data
X_testPred = cModel.predict(X_test)
X_test_AccScore = accuracy_score(X_testPred,Y_test)
```

```python
print('Accuracy with Testing data : ', X_test_AccScore)
```

```python
print(classification_report(Y_test,X_testPred))
```
